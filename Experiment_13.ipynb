```python
import pandas as pd

data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Asha", "Rohan", "Meera", "Karan", "Anil"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "CSE", "cse", "E&TC"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023", "07-06-2023", "10/06/23"]
}

df = pd.DataFrame(data)
df
```

```python
df.replace("-",np.nan, inplace=True)
df
```

```python
#Step-2: Replace '-' with NaN to mark missing values
df.replace("-",np.nan, inplace=True)
#Convert columns to numeric where needed
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df["Marks"]=pd.to_numeric(df["Marks"], errors='coerce')
'''When converting data(like strings -> datatime),
if some values are invalid or cannot be converted,
instead of throwing an error, it will:
-> Replace those invalid values with NaT (Not a Time)'''
df

```

```python
#Fill missing age with mean
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].median(),inplace=True)
df
```

```python
#Step 5: Fix inconsistent text formatting
df["Department"]=df["Department"].str.upper()
df
```

```python
#Step :6 Convert Admission_Date to standard datetime format
df["Admission_Date"]=pd.to_datetime(df["Admission_Date"], format="mixed",dayfirst=True)
df
```
